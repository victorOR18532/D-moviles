**Épsilon**

es el número positivo más pequeño que, al sumarse a 1,produce un resultado diferente de 1
en el sistema de coma flotante de un computador **(1 + e diferente 1)**. Representa el límite superior del error relativo por redondeo, definiendo la precisión máxima de los cálculos numéricos.

```python
epsilon = 1.0

while (1.0 + epsilon) != 1.0:
    epsilon_anterior = epsilon
    epsilon = epsilon / 2.0

print("Épsilon de máquina calculado:", epsilon_anterior)
```

```python
import cmath  # Para manejar números complejos

a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))
c = float(input("Ingrese el valor de c: "))

if a == 0:
    print("No es una ecuación cuadrática.")
else:
    discriminante = b**2 - 4*a*c

    print("Discriminante:", discriminante)

    if discriminante > 0:
        x1 = (-b + discriminante**0.5) / (2*a)
        x2 = (-b - discriminante**0.5) / (2*a)
        print("Dos soluciones reales diferentes:")
        print("x1 =", x1)
        print("x2 =", x2)

    elif discriminante == 0:
        x = -b / (2*a)
        print("Una solución real doble:")
        print("x =", x)

    else:
        x1 = (-b + cmath.sqrt(discriminante)) / (2*a)
        x2 = (-b - cmath.sqrt(discriminante)) / (2*a)
        print("Dos soluciones complejas:")
        print("x1 =", x1)
        print("x2 =", x2)
```
